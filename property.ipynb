```python
#analyze
import pandas as pd
import seaborn as sns
import numpy as np

#visualize
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#store plot in notebook
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read
df = pd.read_csv(r'/Users/fatma/Desktop/excel/property.csv')
```

```python
df.head()
```

```python
#columns type
df.dtypes.head()
```

```python
#convert type

df['price'] = df['price'].astype('Int64')

#df['zipcode'] = df['zipcode'].astype('Int64')
```

```python
#sort
df = df.sort_values(by=['price'], inplace=False, ascending=True)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
df1 = df.groupby('zipcode', sort=False)[['zipcode','price']].mean()
df1.head()
```

```python
#bar chart price and zipcode
x = df1['price']
y = df1['zipcode'].astype('bytes')
fig = plt.figure(figsize = (35, 15))
plt.bar(x, y, edgecolor='black',color='red', width=5)

plt.title("price vs zipcode",  fontsize=30)

plt.xlabel('price' ,fontsize=30)
plt.ylabel('zipcode',  fontsize=30)


plt.show()

```
