```python
# import libraries

#used to analyze data
import pandas as pd 
#use matplotlib to plot graphs, to visualize random distrubution
import seaborn as sns 
import numpy as np

#create visualization
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#make plot outputs appear and be stored within the notebook.
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read in the data df = data frame
df = pd.read_csv(r'/Users/fatma/Desktop/movies.csv')

```

```python
# let's look at the date 

df.head()
```

```python
#let's see if there is any missing data
#
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
                          
```

```python
#data types for columns

df.dtypes
```

```python
#specify column to change data type of columns

df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')
```

```python
df.head()
```

```python
#released coulmn, change type to string andtake the first four
#create correct year coumn

df['yearcorrect'] = pd.to_datetime(df['released'].str.split(' \(').str[0])
df['yearcorrect'] = df['yearcorrect'].astype(str).str[:4]
```

```python
df.head()
```

```python
 # sorting

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# make all column that are object into numberic 
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'): # if the column is string
        df_numerized[col_name] = df_numerized[col_name].astype('category')#change it to category type
        df_numerized[col_name] = df_numerized[col_name].cat.codes#give it random numerazation

df_numerized.head()
```
